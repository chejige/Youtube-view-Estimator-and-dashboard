```python
import pandas as pd
import sqlite3
from datetime import date
```

```python
#connecting with the database.
db = sqlite3.connect("youtbe_video.db")
df.to_sql('youtube_video',db,if_exists='replace',index=False)
print("Successful created!")
```

```python
pd.read_sql(sql,db)
```

```python
link = 'https://drive.google.com/file/d/1-Otig2u-fyeg5MsERCUvfRrgPIX9bHP7/view?usp=sharing'
 
 
# to get the id part of the file
id = link.split("/")[-2]
 
downloaded = drive.CreateFile({'id':id})
downloaded.GetContentFile('video group.csv') 
import pandas as pd
playlist = pd.read_csv('video group.csv',index_col=0)
playlist.head()
```

```python
playlist.to_sql('playlsit',db,if_exists='replace',index=False)
print("Successful created!")
```

```python
sql='''select * from playlsit '''
pd.read_sql(sql,db)
```

```python
# join the two table together based on video ID
sql='''SELECT * FROM youtube_video left outer join playlsit
    ON youtube_video.ID = playlsit.ID'''
df_join=pd.read_sql(sql,db)
```
